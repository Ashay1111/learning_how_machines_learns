# Polynomial Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('../Datasets/Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

## Training the Linear Regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

## Visualising the Simple Linear Regression results

```python
plt.scatter(X,y, marker = 'o', label = 'original data', color = 'red')
plt.plot(X,lin_reg.predict(X), marker = 'x', label = 'predicted data')
plt.legend()
plt.title("Linear Regression")
plt.xlabel('Experience (years)')
plt.ylabel('salary ($)')
plt.show()
```

## Training and Visualising the Polynomial Regression model on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures

#get the regressor
pr = PolynomialFeatures(degree = 2)
polyX = pr.fit_transform(X)

#Train the regressor
poly_reg = LinearRegression()
poly_reg.fit(polyX, y)

#visualize
plt.scatter(X,y, marker = 'o', label = 'original data', color = 'red')
plt.plot(X,poly_reg.predict(polyX), marker = 'x', label = 'predicted data')
plt.legend()
plt.title("Polynomial Regression (Degree : 2)")
plt.xlabel('Experience (years)')
plt.ylabel('salary ($)')
plt.show()
```

```python
#get the regressor
pr = PolynomialFeatures(degree = 4)
polyX = pr.fit_transform(X)

#Train the regressor
poly_reg = LinearRegression()
poly_reg.fit(polyX, y)

#visualize
plt.scatter(X,y, marker = 'o', label = 'original data', color = 'red')
plt.plot(X,poly_reg.predict(polyX), marker = 'x', label = 'predicted data')
plt.legend()
plt.title("Polynomial Regression (Degree : 4)")
plt.xlabel('Experience (years)')
plt.ylabel('salary ($)')
plt.show()
```

## Predicting a new result with Linear Regression

```python
lin_reg.predict([[6.5]]) #should be $160,000
```

## Predicting a new result with Polynomial Regression

```python
poly_reg.predict(pr.fit_transform([[6.51]])) #should be $160,000 
```

Here, Polynomial Regression gives more accurate results.

# Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))

plt.scatter(X, y, color = 'red')
plt.plot(X_grid, poly_reg.predict(pr.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
